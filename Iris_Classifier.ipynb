```python
#Importing Important Libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import keras
```

```python
# Making accessible to google drive

from google.colab import drive

drive.mount('/content/drive')
```

```python
# Reading the dataset

df = pd.read_csv('/content/drive/My Drive/IRIS.csv')
df.head()
```

```python
# Checking basic info of dataset

print(df.shape)
df.describe()
```

```python
# Using sklearn library to label species

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
df.head()
```

```python
# Number of classes: 0 = Iris-setosa, 1 = Iris-versicolor, 2 = Iris-virginica

species_name = le.classes_
print(species_name)
```

```python
# Splitting the dataset into feature and predictors

predictors = df.drop(columns=["species"])
target = df["species"]

print(target.head(3))
predictors.head(3)
```

```python

# Splitting Data into train (70%) and test (30%) dataset

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, shuffle = True, random_state=70)
print(X_train.shape)
print(X_test.shape)
```

```python
# Scaling of Dataset using sklearn

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)           # It alreay Fit
```

```python
# Converting y_train into 2D Binary Matrix

y_train = keras.utils.to_categorical(y_train, num_classes = 3)
print(y_train[:5])
```

```python
# Importing Sequential and Dense library from keras

from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
# Creating Model

model = Sequential()
model.add(Dense(units = 40, activation = 'relu', input_shape = (4,)))
model.add(Dense(units = 40, activation = 'relu'))
model.add(Dropout(0.45))      # ---------------------------------------------------->> Helps to avoid overfitting
model.add(Dense(units = 3, activation = 'softmax'))
```

```python
# Compiling Model

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
# Model Fitting

model.fit(X_train, y_train, epochs = 100, verbose = 1)
```

```python
# Making Predictions

predictions = model.predict(X_test)
print(predictions[:5])
```

```python
# Converting result to more visible form

prediction = np.argmax(predictions, axis = -1)
print(prediction[:5])
```

```python
# Cross Checking

print(y_test[:5])
```

```python
# Checking the Acurracy of Model

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```
